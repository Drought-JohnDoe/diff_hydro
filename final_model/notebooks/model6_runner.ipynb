# Final Model 6 Runner

This notebook gives a very small workflow to:

- edit hyperparameters in a JSON config
- resume training from epoch 20 to 25
- test the new checkpoint
- analyze it
- regenerate the HBV10 comparison report
- print median NSE


```python
from pathlib import Path
import json, subprocess, sys

ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
CONFIG = ROOT / 'configs' / 'model6_resume_ep20_to25.json'
RUNNER = ROOT / 'run_model6.py'
ROOT, CONFIG, RUNNER
```

```python
cfg = json.loads(CONFIG.read_text())
cfg
```

Edit the values below before running if you want to try a different configuration.

Important: for true checkpoint resume, do not change architecture-defining settings like `hidden_size`, `nmul`, or dynamic/static structure. Safe first things to change are usually `max_iter_ep`, `batch_size`, or regularization weights.

```python
cfg['start_epoch'] = 20
cfg['end_epoch'] = 25
cfg['epoch'] = 25
# Example hyperparameter tweaks:
# cfg['max_iter_ep'] = 250
# cfg['reg_amp_w'] = 5e-4
# cfg['reg_smooth_w'] = 5e-4
CONFIG.write_text(json.dumps(cfg, indent=2))
print(CONFIG.read_text())
```

```python
subprocess.run([sys.executable, str(RUNNER), 'resume', '--config', str(CONFIG)], cwd=ROOT, check=True)
```

```python
subprocess.run([sys.executable, str(RUNNER), 'test', '--config', str(CONFIG)], cwd=ROOT, check=True)
subprocess.run([sys.executable, str(RUNNER), 'analyze', '--config', str(CONFIG)], cwd=ROOT, check=True)
subprocess.run([sys.executable, str(RUNNER), 'report', '--config', str(CONFIG)], cwd=ROOT, check=True)
```

```python
subprocess.run([sys.executable, str(RUNNER), 'nse', '--config', str(CONFIG)], cwd=ROOT, check=True)
```
